```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset from the provided URL
url = "https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv"
application_df = pd.read_csv(url)

# Display the first few rows
application_df.head()
```

```python
application_df = application_df.drop(columns=["EIN", "NAME"])
```

```python
# Define the target and feature columns
y = application_df["IS_SUCCESSFUL"]
X = application_df.drop(columns=["IS_SUCCESSFUL"])
```

```python
# Number of unique values per column
unique_counts = X.nunique()
print(unique_counts)

# Show counts for columns with more than 10 unique values
for column in X.columns:
    if X[column].nunique() > 10:
        print(f"\nValue counts for {column}:\n", X[column].value_counts())

```

```python
# Application Type cleanup
app_type_counts = application_df['APPLICATION_TYPE'].value_counts()
rare_apps = app_type_counts[app_type_counts < 500].index
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(rare_apps, "Other")

# Classification cleanup
class_counts = application_df['CLASSIFICATION'].value_counts()
rare_classes = class_counts[class_counts < 1000].index
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(rare_classes, "Other")
```

```python
# Convert categorical variables to dummy/indicator variables
application_df = pd.get_dummies(application_df)
```

```python
# Redefine X and y after encoding
y = application_df["IS_SUCCESSFUL"]
X = application_df.drop(columns=["IS_SUCCESSFUL"])

# Split into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit on training data, transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
